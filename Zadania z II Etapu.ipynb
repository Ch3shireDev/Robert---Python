```python
import sys
import os
from pathlib import Path
from glob import glob
dir1 = ".\\Logia\\etap2\\*\\*png"
_files = glob(dir1)
files = []
for eachfiles in _files:
    f = os.path.abspath(eachfiles)
    f = f[49:]
    files += [f]
    
import re
    
year = ""
num = 0
    
for f in files:
    x = re.search("([0-9]{4})",f).group(1)
    if x != year:
        year = x
        num += 1
        print("# Logia %02d (%d)" % (num, int(year)))
    s = "![](.\\"+f+")"
    s = s.replace('\\', '/')
    print(s)
```
